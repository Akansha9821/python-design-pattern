### It is a type of creational pattern and involves only one class to create methods and specified objects.
### It provides a global point of access to the instance created.

### Real world Example of singleton pattern is python shallow copy

```python
class SingleTon(object):
#   __new__ method will be called when an object is created
    def __new__(cls, *args, **kwargs):
#         hasattr is used to check if an object has the given named attribute and return true if present, else false.
        if not hasattr(cls, '_instance'):
#             super () function is used for accessing the methods and properties of the base class or superclass.
            cls._instance = super().__new__(cls, *args, **kwargs)
        return cls._instance

o1 = SingleTon()
o1.data = 10
print(o1)
print('*************', o1.data)

o2 = SingleTon()
o2.data = 5




print(o2)
print(o1.data)
print(o2.data)
```
